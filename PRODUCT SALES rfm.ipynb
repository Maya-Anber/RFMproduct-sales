```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#group by age, product and gender
rfm = df.groupby(['Age_Group', 'Product_Category','Customer_Gender']).agg(
    {'Date': lambda date:(present - date.max()).days,
     'Order_Quantity': lambda num: len(num), 
     'Revenue': lambda price: price.sum()})
```

```python
rfm.head()
```

```python
# Change the name of columns
rfm.columns =['recency','frequency', 'monetary']
rfm['recency'] = rfm['recency'].astype(int)
rfm.head()
```

```python
#Computing Quartiles of RFM values
rfm['r_quartile'] = pd.qcut(rfm['recency'], 4, ['1','2','3','4'])
rfm['f_quartile'] = pd.qcut(rfm['frequency'], 4, ['4','3','2','1'])
rfm['m_quartile'] = pd.qcut(rfm['monetary'], 4, ['4','3','2','1'])
rfm.head()
```

```python
rfm['RFM_Score'] = rfm.r_quartile.astype(str)+rfm.f_quartile.astype(str) + rfm.m_quartile.astype(str) 
rfm.head()
```

```python
# Filter out Top/Best customers
print(rfm[rfm['RFM_Score']=='111'].sort_values('monetary',ascending =False).head())
print(rfm.columns)
```
